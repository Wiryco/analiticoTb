##### Importação de bibliotecas

```python
import pandas as pd
import itertools
```

##### Importação das bases de dados que estão em CSV

```python
municipio = pd.read_csv(filepath_or_buffer='./dados/d_municipio.csv')
tempo = pd.read_csv(filepath_or_buffer='./dados/d_tempo.csv')
origem = pd.read_csv(filepath_or_buffer='./dados/origem.csv')
```

##### Entendimento da base de dados

```python
display(origem.head(1)) #27820
display(municipio.head(5))
display(tempo.head(1))
```

```python
display(origem.query(expr="co_municipio_residencia_atual == 522140").head(1))
display(municipio.query(expr="dmun_codibge == 522140").head(1))
display(tempo.query(expr="dte_id == 20180824").head(1))
```

##### Unificação das informações usando a função merge do Python

```python
dbaMerge = pd.merge(left=origem,
                        right=municipio,
                        left_on='co_municipio_residencia_atual',
                        right_on='dmun_codibge',
                        how='left',
                        indicator=True)
```

##### Ajusta os dados para fazer um novo merge com a d_tempo

```python
dbaMerge['dataFormatada'] = pd.to_datetime(dbaMerge['dt_diagnostico_sintoma']).dt.strftime('%Y%m%d').astype(int)
```

```python
display(dbaMerge[dbaMerge['dataFormatada'].isna()])
```

##### Filtrar somente o estado de Goiás para começar a parte analítica dos dados.
- Nesse arquivo, deve ser retornada a quantidade de casos novos de tuberculose pulmonar de forma mensal para cada município de Goiás.

```python
dbaFull = pd.merge(left=dbaMerge,
                        right=tempo,
                        left_on='dataFormatada',
                        right_on='dte_id',
                        how='left').query(expr="dmun_uf_sigla in ['GO']")
```

```python
display(dbaFull[dbaFull['month_name'].isna()])
```

```python
dbaFull['month_name'] = dbaFull['month_name'].fillna('N/A').astype(str)
```

##### Aplicando condicionais mediante dicionário de dados informado no repositório.
- Uma coluna chamada município, na qual é esperado que contenha o nome de cada município por extenso.

```python
# Tratativo tp_entrada
dbaFull['motivoEntrada'] = dbaFull.apply(lambda row:
                                         'Caso Novo' if row['tp_entrada'] == 1 else
                                         'Recidiva' if row['tp_entrada'] == 2 else
                                         'Reingresso após Abandono' if row['tp_entrada'] == 3 else
                                         'Não Sabe' if row['tp_entrada'] == 4 else
                                         'Transferência' if row['tp_entrada'] == 5 else
                                         'Pós-óbito' if row['tp_entrada'] == 6 else 'Não informado'
                                         , axis=1)

# Tratativa tp_pop_liberdade
dbaFull['popLiber'] = dbaFull.apply(lambda row:
                                         'Sim' if row['tp_pop_liberdade'] == 1 else
                                         'Não' if row['tp_pop_liberdade'] == 2 else
                                         'Ignorado' if row['tp_pop_liberdade'] == 9 else 'Não informado'
                                         , axis=1)

# Tratativa tp_pop_rua
dbaFull['popRua'] = dbaFull.apply(lambda row:
                                         'Sim' if row['tp_pop_rua'] == 1 else
                                         'Não' if row['tp_pop_rua'] == 2 else
                                         'Ignorado' if row['tp_pop_rua'] == 9 else 'Não informado'
                                         , axis=1)

# Tratativa tp_forma
dbaFull['formaTb'] = dbaFull.apply(lambda row:
                                         'Pulmonar' if row['tp_forma'] == 1 else
                                         'Extrapulmonar' if row['tp_forma'] == 2 else
                                         'Pulmonar + Extrapulmonar' if row['tp_forma'] == 3 else 'Não informado'
                                         , axis=1)

# Tratativa tp_situacao_encerramento
dbaFull['situaEnce'] = dbaFull.apply(lambda row:
                                         'Cura' if row['tp_situacao_encerramento'] == 1 else
                                         'Abandono' if row['tp_situacao_encerramento'] == 2 else
                                         'Óbito por TB' if row['tp_situacao_encerramento'] == 3 else
                                         'Óbito por outras causas' if row['tp_situacao_encerramento'] == 4 else
                                         'Transferência' if row['tp_situacao_encerramento'] == 5 else
                                         'Mudança de Diagnóstico' if row['tp_situacao_encerramento'] == 6 else
                                         'TB-DR' if row['tp_situacao_encerramento'] == 7 else
                                         'Mudança de Esquema' if row['tp_situacao_encerramento'] == 8 else 'Não informado'
                                         , axis=1)

dbaFull['ano'] = pd.to_datetime(dbaFull['dt_diagnostico_sintoma']).dt.strftime('%Y').astype(int)
dbaFull['mes'] = dbaFull.apply(lambda row: str(row['month_name']), axis=1).astype(str)

dbaFull['dtNascimento'] = pd.to_datetime(dbaFull['dt_nascimento']).dt.strftime('%Y-%m-%d')
dbaFull['dtDiagnostico'] = pd.to_datetime(dbaFull['dt_diagnostico_sintoma']).dt.strftime('%Y-%m-%d')

# dbaFull['ref'] = pd.to_datetime(dbaFull['dt_diagnostico_sintoma']).dt.strftime('%Y/%m')
# dbaFull['refExt'] = dbaFull.apply(lambda row: str(row['month_name']) + ' de ' + str(row['ano']), axis=1)

dbaFull['month_number'] = dbaFull['month_number'].fillna(0)

dbaFull.rename(columns={'dmun_municipio': 'municipio',
                        'dmun_uf_sigla': 'uf', 
                        'dmun_uf_nome': 'uf_nome',
                        'dmun_longitude_mun': 'longitude',
                        'dmun_latitude_mun': 'latitude'
                        }, inplace=True)
```

##### Filtra apenas algumas colunas do dataset

```python
filtro = ['nu_notificacao', 'tp_entrada', 'tp_pop_rua', 'tp_forma', 'tp_situacao_encerramento', 'dtNascimento', 'dtDiagnostico', 'motivoEntrada', 'popLiber', 'popRua', 'formaTb', 'situaEnce',
          'uf', 'uf_nome', 'municipio', 'latitude', 'longitude', 'ano', 'mes']

dbaFilter = dbaFull[filtro]
```

##### Aplica condicional:
- Apenas os pacientes que cuja o tipo de entrada em uma unidade seja classificado como "Caso novo", ou seja, todos os pacientes que nunca se submeteram ao tratamento de tuberculose. Ou então, aqueles pacientes cujo tipo de entrada é classificado como "Pós-óbito" ou cujo tipo de entrada é desconhecido.

- Cuja forma clínica de tuberculose é classificada como "Pulmonar" ou "Pulmonar + extrapulmonar".

- Podendo ser qualquer "Situação de encerramento" do caso notificado, exceto "Mudança de Diagnóstico".

- Apenas dados a partir de 2021.

```python
condicional = dbaFilter.query("tp_entrada in [1, 4, 6] and tp_forma in [1, 3] and tp_situacao_encerramento not in [6] and ano > 2021")
```

```python
display(dbaFilter, condicional)
```

```python
gbDf = condicional.groupby(['municipio', 'mes', 'ano']).size().reset_index(name='quantidade')
len(gbDf)
```

##### Montando um dataframe auxiliar para tratar o ponto diferencial do projeto
- Será um diferencial caso consiga retornar para todos os períodos a partir de 2021, todos os 246 municípios de Goiás. Caso algum período para determinado município não tenha valores para serem apresentados, a coluna "Quantidade" deve retornar 0 (zero).

```python
municipios = dbaFull['municipio'].unique()
mes = dbaFull['mes'].unique()
ano = dbaFull['ano'].unique()

print(len(municipios))

distinctMunicipio = pd.DataFrame(list(itertools.product(municipios, mes, ano)), columns=['municipio', 'mes', 'ano']).reset_index(drop=True)
display(distinctMunicipio)
```

```python
display(gbDf)
```

##### Aplicando left join para trazer todos os dados montados com os dois dataframes: o df com o condicional do projeto e o outro df com todos os municípios do estado de Goiás. Quem estiver com a quantidade nula, significa que para aquele período, o estado X não recebeu nenhuma notificação de paciente com Tuberculose. Nesse caso, o contador deve ser 0.

```python
dfEnd = pd.merge(distinctMunicipio, gbDf, on=['municipio', 'mes', 'ano'], how='left').query(expr="ano > 2021")
dfEnd['quantidade'] = dfEnd['quantidade'].fillna(0).astype(int)
display(dfEnd)
```

##### Validando a quantidade de estados que está no dataframe. Tem que ser igual a 247

```python
distinctMunicipioTeste = dfEnd['municipio'].drop_duplicates()

display(distinctMunicipioTeste)
```

##### Criando o campo período concatenando o mes e o ano da notificação.
- Uma coluna chamada mês com o período por extenso no seguinte formato: "<Mês> de <Ano>", exemplo: "Janeiro de 2024".

```python
dfEnd['periodo'] = dfEnd['mes'].astype(str) + ' de ' + dfEnd['ano'].astype(str)
```

```python
display(dfEnd['mes'].drop_duplicates())
```

```python
dfEnd = dfEnd.sort_values(by=['ano', 'mes', 'municipio', 'quantidade'], ascending=True)
dfEnd = dfEnd[['municipio', 'periodo', 'quantidade']]
display(dfEnd)
```

##### Eliminando período onde o mes é N/A (Não existe)

```python
dfEnd = dfEnd[~dfEnd['periodo'].str.contains("N/A")]
```
